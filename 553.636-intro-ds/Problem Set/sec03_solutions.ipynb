```python
import pandas as pd
import numpy as np
# Download the diabetes file, and save it in the working directory
# Read the text file into a pandas dataframe
diabetes_df = pd.read_csv("diabetes.txt", sep='\t')

#1. Y is the output variable for the dataframe. Y is a measure of how far the disease has progressed in 1 year after the initial data collection.

columns = set(diabetes_df.columns)
columns.remove("Y")
for column in columns:
    col_values = np.array(diabetes_df[column])
    # calculate the mean and std for each attribute
    print(column, "\n std:", np.std(col_values), "\n mean:", np.mean(col_values))
    # calculate the standard deviation
    np.std(col_values)
    
# calculate the correlation coefficient matrix
print("Correlation coefficient matrix (computed with numpy):") 
display(pd.DataFrame(np.corrcoef(np.array(diabetes_df).T),columns = diabetes_df.columns))

print("Correlation coefficient matrix (computed with pandas):")
display(diabetes_df.corr())
```

The main diagonal of the correlation coefficient is equal to a vector of ones because every variable has full correlation with itself. 

3. It does not truly make sense to calculate the mean or standard deviation for the SEX column. It is a categorical variable, and thus mean does not make sense. Furthermore, standard deviation is not useful, since there are no values besides 1 and 2.

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

# It does not make sense to use linear regression on the sex column, so we drop it
lin_diab_view = diabetes_df.drop(columns = ['SEX'])

# normalize the values, 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
saved_cols = lin_diab_view.columns
lin_diab_view_arr = scaler.fit_transform(lin_diab_view)
display(pd.DataFrame(lin_diab_view_arr, columns = saved_cols))
#lin_diab_view.apply(lambda x: x/x.max() - x.mean()/x.max(), axis=0)

# separate entries into train and test segments
from sklearn.model_selection import train_test_split

train, test = train_test_split(lin_diab_view, test_size=0.2)
train_X = train.drop(columns = ["Y"])
train_y = train["Y"]
test_X = test.drop(columns = ["Y"])
test_y = test["Y"]

# Create linear model and train and test for two columns
from sklearn.linear_model import LinearRegression
diab_pred = LinearRegression().fit(train_X[["AGE", "BP"]], train_y)
print("Result for Age and Blood Pressure", diab_pred.score(test_X[["AGE", "BP"]], test_y))

# Test for all columns
diab_pred = LinearRegression().fit(train_X, train_y)
print("Result for regressing on all columns",diab_pred.score(test_X, test_y))
```

Age and BP is the best I've found so far. There are many option that are better than using all of them, though. Age, BMI, and Age, BP are just two of them.
